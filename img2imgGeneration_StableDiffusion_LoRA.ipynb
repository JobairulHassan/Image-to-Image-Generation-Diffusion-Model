Prepare Dataset using Custom Class

```python
from torch.utils.data import DataLoader, Dataset
import os
from PIL import Image
from torchvision import transforms
import torch.optim as optim

# Data augmentation and transformation
transform = transforms.Compose([
    transforms.Resize((512, 512)),
    transforms.ToTensor(),
])

class ImageToImageDataset(Dataset):
    def __init__(self, input_dir, target_dir, prompt_dir, transform):
        self.input_images = sorted(os.listdir(input_dir))
        self.target_images = sorted(os.listdir(target_dir))
        self.prompt_files = sorted(os.listdir(prompt_dir))  # Assuming prompts are in text files
        self.input_dir = input_dir
        self.target_dir = target_dir
        self.prompt_dir = prompt_dir
        self.transform = transform

    def __len__(self):
        return len(self.input_images)

    def __getitem__(self, idx):
        # Paths for images and prompt
        input_image_path = os.path.join(self.input_dir, self.input_images[idx])
        target_image_path = os.path.join(self.target_dir, self.target_images[idx])
        prompt_path = os.path.join(self.prompt_dir, self.prompt_files[idx])  # Get prompt corresponding to the image

        # Load images
        input_image = Image.open(input_image_path).convert("RGB")
        target_image = Image.open(target_image_path).convert("RGB")

        # Load and read the prompt text
        with open(prompt_path, 'r') as file:
            prompt = file.read().strip()  # Remove any leading/trailing whitespace

        # Apply transformations to the images
        input_image = self.transform(input_image).unsqueeze(0).to("cuda")
        target_image = self.transform(target_image).unsqueeze(0).to("cuda")

        # Return the image pair along with the prompt
        return input_image, target_image, prompt

# Example dataset directories
input_dir = "/content/drive/MyDrive/Colab Notebooks/dataset/train/source"
target_dir = "/content/drive/MyDrive/Colab Notebooks/dataset/train/target"
prompt_dir = "/content/drive/MyDrive/Colab Notebooks/dataset/train/prompts"  # Path to your prompts directory

dataset = ImageToImageDataset(input_dir, target_dir, prompt_dir, transform)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)
```

**Train Model**

```python
import torch
import torch.optim as optim
import torch.nn as nn
from torchvision import transforms
from diffusers import StableDiffusionPipeline

# Define loss function and optimizer
loss_fn = nn.MSELoss()
optimizer = optim.AdamW(pipe.unet.parameters(), lr=1e-5)

# Training loop
def train_model(dataloader, pipe, optimizer, loss_fn, epochs=5):
    pipe.unet.train()  # Ensure UNet is in training mode
    to_tensor = transforms.ToTensor()

    for epoch in range(epochs):
        total_loss = 0

        for input_image, target_image, prompt in dataloader:
            # Send images to device (GPU)
            input_image = input_image.to("cuda")
            target_image = target_image.to("cuda")

            optimizer.zero_grad()

            # Forward pass using the pipeline with original prompt from the dataset
            output = pipe(prompt=prompt[0], init_image=input_image, strength=0.75)  # Use the first prompt in the batch

            generated_image = output.images[0]

            # Convert generated image to tensor
            generated_image_tensor = to_tensor(generated_image).unsqueeze(0).to(target_image.device)

            # Calculate loss between generated and target image
            loss = loss_fn(generated_image_tensor, target_image)
            loss.requires_grad = True
            loss.backward()  # Backpropagate the loss

            optimizer.step()  # Update model parameters

            total_loss += loss.item()

        print(f"Epoch {epoch + 1}/{epochs}, Loss: {total_loss / len(dataloader)}")

# Fine-tune the model
train_model(dataloader, pipe, optimizer, loss_fn, epochs=10)
```

**Run on test data**

```python
import safetensors.torch
# Paths for test data
input_dir = "/content/drive/MyDrive/Colab Notebooks/dataset/val/source"
target_dir = "/content/drive/MyDrive/Colab Notebooks/dataset/val/target"
prompt_dir = "/content/drive/MyDrive/Colab Notebooks/dataset/val/prompts"
output_dir = "/content/drive/MyDrive/Colab Notebooks/dataset/val/generated_output"  # Directory to save generated images

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Initialize the dataset and dataloader for test data
test_dataset = ImageToImageDataset(input_dir, target_dir, prompt_dir, transform)
test_dataloader = DataLoader(test_dataset, batch_size=4, shuffle=False)

# Load the fine-tuned model (Stable Diffusion pipeline)
unet_state_dict = safetensors.torch.load_file("/content/save_model/diffusion_pytorch_model.safetensors")
pipe.unet.load_state_dict(unet_state_dict)

# Set model to evaluation mode
pipe.unet.eval()

# Function to test the model and save the outputs
def test_model(dataloader, pipe, output_dir):
    to_tensor = transforms.ToTensor()

    with torch.no_grad():  # No gradients needed during inference
        for idx, (input_image, target_image, prompt) in enumerate(dataloader):
            # Send images to device (GPU)
            input_image = input_image.to("cuda")
            target_image = target_image.to("cuda")

            # Generate image using the pipeline
            output = pipe(prompt=prompt[0], init_image=input_image, strength=0.75)  # Using the first prompt in the batch
            generated_image = output.images[0]

            # Save the generated image to the output directory
            output_image_path = os.path.join(output_dir, f"generated_image_{idx+1}.png")
            generated_image.save(output_image_path)
            print(f"Saved image {idx+1} to {output_image_path}")

# Run the test on your test dataset and save the results
test_model(test_dataloader, pipe, output_dir)
```
